```python
df3= df2.copy()
df3['Price_per_sqft'] = df3['Price']/df3['Area']
```

```python
len(df3.Location.unique())
```

```python
df3.Location = df3.Location.apply(lambda x: x.strip())
location_stats = df3.groupby('Location')['Location'].agg('count').sort_values(ascending=False)
print(location_stats)
```

```python
len(location_stats[location_stats<=5])
```

```python
locations_Under_5 = location_stats[location_stats<=5]
print(locations_Under_5)
```

```python
df3.Location = df3.Location.apply(lambda x:'other' if x in locations_Under_5 else x)
len(df3.Location.unique())
```

```python
df3.head()
```

```python
df4 = df3.copy()
df4.rename( columns=({ 'No. of Bedrooms': 'bhk','Lift Available':'Lift','Car Parking':'Car','24x7 Security':'Security'}), 
    inplace=True,
)
df4.head()
```

```python
df4[df4.Area/df4.bhk<300].head()
```

```python
df4.shape
```

```python
df5 = df4[~(df4.Area/df4.bhk<300)]
df5.shape
```

```python
df5.Price_per_sqft.describe()
```

```python
df5.head()
```

```python

dummies = pd.get_dummies(df5.Location)
dummies.head(4)
```

```python

df6 = pd.concat([df5,dummies.drop('other',axis='columns')],axis='columns')
df6.head(3)
```

```python
df7 = df6.drop('Location',axis='columns')
df7.head(2)
```

```python
df7.shape
```

```python
X = df7.drop('Price',axis='columns')
X.head()
```

```python
Y1 = df7.Price
Y = Y1/100000
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
lr.score(X_test,Y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)

cross_val_score(LinearRegression(),X,Y,cv=cv)
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(splitter="random")
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
def predict_price1(Location,Area,bhk,Car,Lift):
  loc_index = np.where(X.columns==Location)[0][0]

  x = np.zeros(len(X.columns))
  x[0] = Area
  x[1] = bhk
  x[5] = Car
  x[3] = Lift

  if loc_index >=0:
    x[loc_index] = 1

  return model.predict([x])[0]
```

```python
predict_price1('mumbai',3500,6,2,3)
```
